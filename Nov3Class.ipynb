```python
def PDF(epsilon):
    x1=np.linspace(-3,3,100)
    x2=np.linspace(-3,3,100)
    X1, X2= np.meshgrid(x1, x2)
    exp_arg = -1/(2*(1+epsilon**2)) * (X1**2 - 2*epsilon*X1*X2 + X2**2)
    return 1/(2*np.pi) * np.exp(exp_arg)
    
```

```python
epsilon = 0.01
plt.imshow(PDF(epsilon), extent=[-3,3,-3,3])
```
