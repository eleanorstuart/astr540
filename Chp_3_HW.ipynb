```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
```

```python
zs = np.linspace(0, 6, 100)
```

```python
# Hubble Parameter
def H(z, omega_M = 0, omega_R = 0, omega_lambda = 0):
    H_z = np.sqrt(omega_M * (1+z)**3 + omega_R * (1+z)**4 + omega_lambda)
    return H_z
```

```python
# Comoving Distance
def r(z, omega_M = 0, omega_R = 0, omega_lambda = 0):
    result = quad(lambda x: 1.0/H(x, omega_M=omega_M, omega_R=omega_R, omega_lambda=omega_lambda),
                  0, z )
    return result[0]
```

```python
# Luminosity Distance
def d_L(z, omega_M = 0, omega_R = 0, omega_lambda = 0):
    return (1 + z) * r(z, omega_M=omega_M, omega_R=omega_R, omega_lambda=omega_lambda)
```

```python
d_L_M = []
d_L_lambda = []
d_L_lambdaCDM = []
d_L_R = []
for z in zs:
    d_L_M.append(d_L(z, omega_M=1))
    d_L_lambda.append(d_L(z, omega_lambda=1))
    d_L_lambdaCDM.append(d_L(z, omega_M=0.3, omega_lambda=0.7))
    d_L_R.append(d_L(z, omega_R=1))
```

```python
plt.plot(zs, d_L_M, 'b', label='omega_M = 1')
plt.plot(zs, d_L_lambda, 'c', label='omega_lambda = 1')
plt.plot(zs, d_L_lambdaCDM, 'r', label='lambda CDM')
plt.plot(zs, d_L_R, 'm', label='omega_R = 1')
plt.legend(loc='upper left')
plt.xlabel('Redshift z')
plt.ylabel('Luminosity Distance d_L in 1/H_0')
plt.ylim(0, 18)
plt.title('Luminosity Distance - Redshift relation')
plt.plot()
```

```python
# Angular Diameter Distance
def d_A(z, omega_M = 0, omega_R = 0, omega_lambda = 0):
    return r(z, omega_M=omega_M, omega_R=omega_R, omega_lambda=omega_lambda)/(1+z)
```

```python
d_A_M = []
d_A_lambda = []
d_A_lambdaCDM = []
d_A_R = []
for z in zs:
    d_A_M.append(d_A(z, omega_M=1))
    d_A_lambda.append(d_A(z, omega_lambda=1))
    d_A_lambdaCDM.append(d_A(z, omega_M=0.3, omega_lambda=0.7))
    d_A_R.append(d_A(z, omega_R=1))
```

```python
plt.plot(zs, d_A_M, 'b', label='omega_M = 1')
plt.plot(zs, d_A_lambda, 'c', label='omega_lambda = 1')
plt.plot(zs, d_A_lambdaCDM, 'r', label='lambda CDM')
plt.plot(zs, d_A_R, 'm', label='omega_R = 1')
plt.legend(loc='upper left')
plt.xlabel('Redshift z')
plt.ylabel('Angular Diameter Distance d_A in 1/H_0')
plt.title('Angular Diameter Distance - Redshift relation')
plt.plot()
```
